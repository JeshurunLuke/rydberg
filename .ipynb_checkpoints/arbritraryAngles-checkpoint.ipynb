# Simple Test case w/ B-Field in plane

```python
%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd

from arc import *  # Import ARC (Alkali Rydberg Calculataor)
```

```python
Ectrl = np.linspace(0, 130, 50)
angle_scan = np.zeros(len(Ectrl))
E_scan_total = np.zeros(len(Ectrl))
angle_res = 22
alpha = 0.73
Eres = 104

Eact = lambda Ei, angle_i : np.sqrt((alpha*Ei)**2 + Eres**2 - 2*(alpha*Ei)*Eres*np.cos(angle_i*np.pi/180))
for ind, E_i in enumerate(Ectrl): 
    E_total = Eact(E_i, angle_res)
    xcomp = Eres*np.cos(angle_res*np.pi/180) - E_i*alpha
    ycomp = Eres*np.sin(angle_res*np.pi/180)
    E_res_vec = np.array([xcomp, ycomp])
    angle_scan[ind] = np.arccos(np.dot(np.array([0, 1]), E_res_vec)/np.linalg.norm(E_res_vec))
    E_scan_total[ind] = E_total/10
```

```python
import pickle


calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False)

calc.diagonalise(Earray = E_scan_total,Barray = [1e-4], theta=angle_scan, drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
RES_mol = 2.227945
RETcond = RET(calc, RES_mol)
sols = RETcond.find_resonance(98, 2)
```

```python
plt.clf()
convFac = 1e6
for key in reversed(sols): 
    comp, e_ans = sols[key]
    
    print(comp[len(comp) -1])
    print(e_ans[0, : ])
    print(e_ans[1, : ]*convFac)
    print(Ectrl[np.argmin(np.abs(e_ans[1, : ]*convFac))])
    print(np.argmin(np.abs(e_ans[1, : ]*convFac)))
    plt.plot(Ectrl, (e_ans[2, : ])*convFac)
    
    
plt.show()
```

# Generalized Case

```python
%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd
from scipy.spatial.transform import Rotation as R
from arc import *  # Import ARC (Alkali Rydberg Calculataor)
```

```python
Ectrl = np.linspace(70, 90, 60)
angle_scan_theta = np.zeros(len(Ectrl))
angle_scan_phi = np.zeros(len(Ectrl))
angle_psi = 55*np.pi/180


E_scan_total = np.zeros(len(Ectrl))
angle_res = 22
alpha = 0.73
Eres = 104

Eact = lambda Ei, angle_i : np.sqrt((alpha*Ei)**2 + Eres**2 - 2*(alpha*Ei)*Eres*np.cos(angle_i*np.pi/180))
for ind, E_i in enumerate(Ectrl): 
    E_total = Eact(E_i, angle_res)
    ycomp = Eres*np.cos(angle_res*np.pi/180)
    zcomp = Eres*np.sin(angle_res*np.pi/180)

    
    E_res_vec = np.array([0, ycomp, zcomp])
    rot_axis = np.array([0, 1, 0])
    r = R.from_rotvec(angle_psi * rot_axis)
    E_res_vec = r.as_matrix() @ E_res_vec

    E_Total = E_res_vec -  np.array([0, E_i*alpha, 0])
    
    angle_scan_theta[ind] = np.arccos(np.dot(np.array([0,0, 1]), E_Total)/np.linalg.norm(E_Total))
    proj_plane = np.concatenate((np.array(E_Total[0:2]),[0]))
    proj_plane = proj_plane/np.linalg.norm(proj_plane)
    angle_scan_phi[ind] = np.arccos(np.dot(np.array([1,0, 0]), proj_plane)/np.linalg.norm(proj_plane))

    E_scan_total[ind] = E_total/10
```

```python
import pickle


calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False)

calc.diagonalise(Earray = E_scan_total,Barray = [2e-4], theta=angle_scan_theta,phi = angle_scan_phi,  drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
RES_mol = 2.227945 #+ 0.01
#RES_mol = 2.226
RETcond = RET(calc, RES_mol)
sols = RETcond.find_resonance(98, 2)
```

```python
plt.clf()
convFac = 1e6
debugOutput = False

for key in reversed(sols): 
    comp, e_ans = sols[key]
    if debugOutput: 
        print(comp[len(comp) -1])
        print(e_ans[0, : ])
        print(e_ans[1, : ]*convFac)
        print(Ectrl[np.argmin(np.abs(e_ans[1, : ]*convFac))])
        print(np.argmin(np.abs(e_ans[1, : ]*convFac)))
    plt.plot(Ectrl, (e_ans[2, : ])*convFac)
    break
    
    
plt.show()
```

```python
from scipy.signal import argrelextrema
plt.clf()
for key in reversed(sols):
    comp, e_ans = sols[key]
    spectrum = np.abs(e_ans[1, : ])*convFac
    indsMax = argrelextrema(spectrum, np.greater)
    plt.vlines(x = Ectrl[indsMax], ymin = 0, ymax = spectrum[indsMax],
    )
    plt.scatter(Ectrl[indsMax],spectrum[indsMax])
plt.show()
    
```

# B-Field Calibration

```python
%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd
from scipy.spatial.transform import Rotation as R
from arc import *  # Import ARC (Alkali Rydberg Calculataor)
```

```python
angle_psi = 45


def Configuration(E_ctrl, psi, E_res = 104, angle_res = 22.1, alpha = 0.73): 
    ycomp = E_res*np.cos(angle_res*np.pi/180) - E_ctrl*alpha
    zcomp = E_res*np.sin(angle_res*np.pi/180)
    E_res_vec = np.array([0, ycomp, zcomp])
    rot_axis = np.array([0, 1, 0])
    r = R.from_rotvec(psi*np.pi/180 * rot_axis)
    E_res_vec = r.as_matrix() @ E_res_vec
    angle_theta = np.arccos(np.dot(np.array([0,0, 1]), E_res_vec)/np.linalg.norm(E_res_vec))
    
    proj_plane = np.concatenate((np.array(E_res_vec[0:2]),[0]))
    proj_plane = proj_plane/np.linalg.norm(proj_plane)
    angle_phi =  np.arccos(np.dot(np.array([1,0, 0]), proj_plane)/np.linalg.norm(proj_plane))

    E_total = np.linalg.norm(E_res_vec)/10
    return E_total,angle_theta, angle_phi
    
E_T, angle_theta, angle_phi = Configuration(20, angle_psi)
```

```python

calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False)
B_scan = np.arange(0, 16, 2)*1e-4


calc.diagonalise(Earray = [E_T],Barray = B_scan, theta=angle_theta ,phi = angle_phi,  drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
plt.clf()
calc.plotLevelDiagram(units='ghz', ylim = 0.1, displayAll =False)
calc.showPlot()
```
