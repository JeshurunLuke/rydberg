# Add Resonance Search

```python
%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd

from arc import *  # Import ARC (Alkali Rydberg Calculataor)
```

```python
import pickle


calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False)
Ectrl = np.linspace(50, 110, 40)
alpha = 0.73
angle = 22.6
thetaTest = 50
Eres = 104
Eact = lambda Ei: np.sqrt((alpha*Ei)**2 + Eres**2 - 2*(alpha*Ei)*Eres*np.cos(22*np.pi/180))
Earray = Eact(Ectrl)/10
calc.diagonalise(Earray = Earray,Barray = [2e-4], theta= (thetaTest)*np.pi/180, drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
plt.clf()
calc.plotLevelDiagram(units='ghz', ylim = 0.1, displayAll =False)
calc.showPlot()
```

```python
RES_mol = 2.228
RETcond = RET(calc, RES_mol)
sols = RETcond.find_resonance(98, 2)
```

```python
plt.clf()
debug = True
convFac = 1e6
for key in reversed(sols): 
    comp, e_ans = sols[key]
    if debug: 
        print(comp[-1])
        print(e_ans[0, : ])
        print(e_ans[1, : ]*convFac)
        print(Ectrl[np.argmin(np.abs(e_ans[1, : ]*convFac))])
        print(np.argmin(np.abs(e_ans[1, : ]*convFac)))
    #plt.plot(Ectrl, 1/abs(e_ans[1, : ])*convFac)
    plt.plot(Ectrl, e_ans[2, : ]*convFac)
    break
    #plt.plot(Ectrl, abs(e_ans[0, : ]))
    
    

plt.show()
```
